## **House Price Prediction Tasks:**
### **1. Dataset Loading:**


```python

import pandas as pd

# Load dataset
df = pd.read_csv("/content/Digi- House price prediction dataset.csv")

# Display first 10 rows
print("First 10 Rows:")
print(df.head(10))
```

### **2. Data Exploration:**

```python
# Display the shape of the dataset
print("\nDataset Shape:")
print(df.shape)

# Check data types of all columns
print("\nData Types:")
print(df.dtypes)

# Generate summary statistics of numerical features
print("\nSummary Statistics:")
print(df.describe())
```

### **3. Data Cleaning:**

```python
# Identify missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Handle missing values appropriately

# Fill categorical missing values with mode
df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode()[0])
df['Exterior1st'] = df['Exterior1st'].fillna(df['Exterior1st'].mode()[0])

# Fill numerical missing values with median
df['BsmtFinSF2'] = df['BsmtFinSF2'].fillna(df['BsmtFinSF2'].median())
df['TotalBsmtSF'] = df['TotalBsmtSF'].fillna(df['TotalBsmtSF'].median())

# Remove SalePrice because most values are missing
df = df.drop(columns=["SalePrice"])

# Verify missing values again
print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# Check and remove duplicate records if any
print("\nDuplicate Records:")
print(df.duplicated().sum())
```

### **4. Feature Selection:**

```python
# Identify relevant features for analysis

# Display all features
print("Features before selection:")
print(df.columns)

# Remove unnecessary column
df.drop("Id", axis=1, inplace=True)

# Display selected features
print("\nRelevant Features:")
print(df.columns)
```

### **5. Data Preprocessing:**

```python
# Convert categorical variables into numerical form (encoding)
from sklearn.preprocessing import LabelEncoder
# One-Hot Encoding
df_encoded = pd.get_dummies(df, drop_first=True)

# Preparing the dataset for machine learning and display it’s preview in the output
print("Preprocessed Dataset")
print(df_encoded.head(10))
```
